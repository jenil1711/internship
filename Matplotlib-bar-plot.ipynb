# Bar Graph:  (also called Bar Chart) is a graphical display of data using bars of different heights. 
It is a really good way to show relative sizes.

```python
Grades = ['A' , 'B', 'C', 'D']
Test1=  [3,13,10,4] 
Test2=  [5, 14,  6, 5]
Test3=  [7,  15, 5, 3]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.bar(Grades,Test1)
```

# Give Title, Xlabel, and Ylabel

```python
plt.bar(Grades,Test1)
plt.title("Grades of test 1",fontsize=20)
plt.xlabel("Grades",fontsize=15)
plt.ylabel("No of students",fontsize=15)
```

# Set Width of The Bars & Align Bars To Center

```python
plt.bar(Grades,Test1,width=0.4,align='edge')
plt.title("Grades of test 1")
plt.xlabel("Grades")
plt.ylabel("No of students")
plt.show()
```

# Change Color of The Bars & Also Change Edges Color

```python
plt.bar(Grades,Test1,width=0.4,align='edge',color='g',edgecolor='r')
plt.title("Grades of test 1")
plt.xlabel("Grades")
plt.ylabel("No of students")
plt.show()
```

# Change  Line Styles & Line Width of The Bars

```python
plt.bar(Grades,Test1,width=0.4,align='edge',color='g',edgecolor='r',ls='--',lw=2)
plt.title("Grades of test 1")
plt.xlabel("Grades")
plt.ylabel("No of students")
plt.show()
```

# Draw Bar Plots For Test1 and Test2

```python
index = np.arange(len(Grades))
index
```

```python
plt.bar(index,Test1,width=0.4,align='edge',color='g',edgecolor='r',ls='--',lw=2)
plt.bar(index+0.3,Test2,width=0.4,align='edge',color='b',edgecolor='r',ls='--',lw=2)
plt.title("Grades of test 1")
plt.xlabel("Grades")
plt.ylabel("No of students")
plt.xticks(index+0.2,Grades)
plt.show()
```

# Draw Bar Plots For Test1, Test2 & Test3

```python
plt.bar(index,Test1,width=0.3,align='edge',color='g',edgecolor='r',ls='--',lw=2)

plt.bar(index+0.3,Test2,width=0.3,align='edge',color='b',edgecolor='r',ls='--',lw=2)

plt.bar(index+0.6,Test3,width=0.3,align='edge',color='y',edgecolor='r',ls='--',lw=2)

plt.title("Grades of test 1")
plt.xlabel("Grades")
plt.ylabel("No of students")
plt.xticks(index+0.3,Grades)
plt.show()
```

# Add Legend 

```python
plt.bar(index,Test1,width=0.3,align='edge',color='g',edgecolor='r',ls='--',lw=2,label='Test1')

plt.bar(index+0.3,Test2,width=0.3,align='edge',color='b',edgecolor='r',ls='--',lw=2,label='Test2')

plt.bar(index+0.6,Test3,width=0.3,align='edge',color='y',edgecolor='r',ls='--',lw=2,label='Test3')

plt.legend()

plt.title("Grades of test 1")
plt.xlabel("Grades")
plt.ylabel("No of students")
plt.xticks(index+0.3,Grades,rotation=90)
plt.show()
```

# Change Figure Size & Color

```python
plt.figure(figsize=(10,3),facecolor='lightblue',edgecolor='k',linewidth=2)
plt.bar(index,Test1,width=0.3,align='edge',color='g',edgecolor='r',ls='--',lw=2,label='Test1')

plt.bar(index+0.3,Test2,width=0.3,align='edge',color='b',edgecolor='r',ls='--',lw=2,label='Test2')

plt.bar(index+0.6,Test3,width=0.3,align='edge',color='y',edgecolor='r',ls='--',lw=2,label='Test3')

plt.legend()
plt.title("Grades of test 1")
plt.xlabel("Grades")
plt.ylabel("No of students")
plt.xticks(index+0.3,Grades)
plt.show()
```
