# Data visualization using matplotlib

# Line plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Draw a simple line plot

```python
x =[1,2,4,7,9]
y =[1,2,3,4,5]
plt.plot(x,y)
```

# Give title, xlabel and ylabel

```python
x =[1,2,4,7,9]
y =[1,2,3,4,5]
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

# Change Font Size of Title, Xlabel and Ylabel And Also Change location

```python
x =[1,2,4,7,9]
y =[1,2,3,4,5]
plt.plot(x,y)
plt.title("Line Plot",fontsize=15,loc='left')
plt.xlabel("X-Axis",fontsize=15,loc='right')
plt.ylabel("Y-Axis",fontsize=15,loc='bottom')
```

#  Change Color & Width of The Line

```python
x =[1,2,4,7,9]
y =[1,2,3,4,5]
plt.plot(x,y,color='g',linewidth=4)
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
```

# Change Marker, Marker Size &  Line Style

```python
x =[1,2,4,7,9]
y =[1,2,3,4,5]
plt.plot(x,y,color='g',linewidth=4,marker='s',markersize='6',markerfacecolor='blue',markeredgecolor='blue',linestyle='--')
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
```

# Add Legend To Your Graph Also Change Location

```python
x = np.random.randint(low=1,high=10,size=20)
y = np.random.randint(low=1,high=10,size=20)
print(x)
print(y)
```

```python
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,y,'b--<',linewidth=4,markersize=10,
        markerfacecolor='g',markeredgecolor='blue')
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.legend(["A simple line plot"],loc=0)
```

# Create Separate Lines For x and y And Change Plot Style

```python
#plt.style.available
```

```python
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,'g--<',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b-->',linewidth=4,markersize=10,
        markerfacecolor='y',markeredgecolor='blue',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.legend()
```

# Xlim And Ylim

```python
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,'g--<',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b-->',linewidth=4,markersize=10,
        markerfacecolor='y',markeredgecolor='blue',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')

plt.xlim(0,20)
plt.ylim(0,15)

plt.legend()
```

```python
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,'g--<',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b-->',linewidth=4,markersize=10,
        markerfacecolor='y',markeredgecolor='blue',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')

# xmin xmax ymin ymax:
plt.axis([0,20,0,15])


plt.legend()
```

# Change X and Y limit Using Xticks And Yticks.

```python
x_ticks=[x for x in range(0,20,2)]
y_ticks=[y for y in range(0,15,3)]
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,'g--<',linewidth=4,markersize=10,
        markerfacecolor='red',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b-->',linewidth=4,markersize=10,
        markerfacecolor='y',markeredgecolor='blue',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
```

```python
x_ticks=[x for x in range(0,20,2)]
y_ticks=[y for y in range(0,15,3)]
x=np.random.randint(low=1,high=10,size=20)
y=np.random.randint(low=1,high=8,size=20)
plt.plot(x,'g--s',linewidth=4,markersize=4,
        markerfacecolor='red',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--s',linewidth=4,markersize=4,
        markerfacecolor='y',markeredgecolor='blue',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color='r',linestyle='--')
```

# change figure size
