# Pie Chart: a special chart that uses "pie slices" to show relative sizes of data.

```python
import matplotlib.pyplot as plt
```

```python
movie = ['Comedy','Action','Romance','Drama']
votes=[5,6,3,1]
```

```python
plt.pie(votes,labels=movie)
plt.title("Favorite Type of Movie")
plt.show()
```

```python
plt.pie(votes,labels=movie)
plt.title("Favorite Type of Movie")
plt.legend(loc=1)
plt.show()
```

```python
plt.pie(votes,labels=movie,labeldistance=1.2)
plt.title("Favorite Type of Movie")
plt.legend(loc=1)
plt.show()
```

# Pull out action and drama wedges 

```python
plt.pie(votes,labels=movie,labeldistance=1.2,explode=[0,0.1,0,0.1])
plt.title("Favorite Type of Movie")
plt.legend(loc=1)
plt.show()
```

# change start angle by 90 degree

```python
plt.pie(votes,labels=movie,labeldistance=1.2,explode=[0,0.1,0,0.1],startangle=90)
plt.title("Favorite Type of Movie")
plt.legend(loc=1)
plt.show()
```

```python
plt.pie(votes,labels=movie,labeldistance=1.2,explode=[0,0.1,0,0.1],startangle=90,colors=['y','g','c','m'])
plt.title("Favorite Type of Movie")
plt.legend(loc=1)
plt.show()
```

# display percentage & change distance inside each wedge

```python
plt.pie(votes,labels=movie,labeldistance=1.2,explode=[0,0.1,0,0.1],startangle=0,colors=['y','g','c','m'],autopct='%0.2f%%',pctdistance=0.7)
plt.title("Favorite Type of Movie")
plt.legend(loc=1)
plt.show()
```

# change the radius of pi chart

```python
plt.pie(votes,labels=movie,labeldistance=1.2,
        explode=[0,0.1,0,0.1],startangle=0,
        colors=['y','g','c','m'],radius=1.3,
        autopct='%0.2f%%',pctdistance=0.7)
plt.title("Favorite Type of Movie",y=1.1)
plt.legend(loc=1)
plt.show()
```

```python
plt.pie(votes,labels=movie,labeldistance=1.2,
        explode=[0,0.1,0,0.1],startangle=0,
        colors=['y','g','c','m'],radius=1,
        autopct='%0.2f%%',pctdistance=0.7,
        textprops={'fontsize':12}
       )
plt.title("Favorite Type of Movie",y=1.1)
plt.legend(loc=1)
plt.show()
```

```python
plt.pie(votes,labels=movie,labeldistance=1.2,
        explode=[0,0.1,0,0.1],startangle=0,
        colors=['y','g','c','m'],radius=1,
        autopct='%0.2f%%',pctdistance=0.7,
        textprops={'fontsize':12},
        wedgeprops={'edgecolor':'k','linestyle':'--','linewidth':2},
        shadow=True
       )
plt.title("Favorite Type of Movie",y=1.1)
plt.legend(loc=1)
plt.show()
```

```python
plt.figure(figsize=(10,3),facecolor="lightblue",edgecolor="black",linewidth=3)
plt.pie(votes,labels=movie,labeldistance=1.2,
        explode=[0,0.1,0,0.1],startangle=0,
        colors=['y','g','c','m'],radius=1,
        autopct='%0.2f%%',pctdistance=0.7,
        textprops={'fontsize':12},
        wedgeprops={'edgecolor':'k','linestyle':'--'},
        shadow=True
       )
plt.title("Favorite Type of Movie",y=1.1)
plt.legend(loc=1)
plt.show()
```
