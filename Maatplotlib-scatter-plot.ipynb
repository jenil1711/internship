# Scatter Plot

```python
temp= [14.2,16.4,11.9,15.2,18.5,22.1,19.4]
icecream_sales=[215,325,185,332,406,522,412]
tea_sales=[80,65,89,70,56,45,43]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(temp,icecream_sales)
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")

plt.show()
```

```python
plt.scatter(temp,icecream_sales,s=100)
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")

plt.show()
```

```python
plt.scatter(temp,icecream_sales,c='y')
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")

plt.show()
```

# change the marker

```python
plt.scatter(temp,icecream_sales,c='y',marker='<')
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")

plt.show()
```

# Change The Edge Color of The Marker Also change Line Width

```python
plt.scatter(temp,icecream_sales,c='y',marker='<',edgecolor='k',linewidth=2)
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")

plt.show()
```

# Change Alpha Parameter
for transparncy of graph


```python
plt.scatter(temp,icecream_sales,c='y',alpha=0.88)
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")

plt.show()
```

#  Set Different Colors For Different Dots

```python
color = [x for x in range (11,18)]
color
```

```python
plt.scatter(temp,icecream_sales,c=color,marker='<')
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")

plt.show()
```

```python
plt.scatter(temp,icecream_sales,c=color,marker='<',cmap="Oranges")
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")

plt.show()
```

# Add colorbar

```python
plt.scatter(temp,icecream_sales,c=color,marker='<',cmap="Oranges")
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")
plt.colorbar()
plt.show()
```

```python
plt.scatter(temp,icecream_sales,c=color,marker='<',cmap="Oranges",vmin=10,vmax=20)
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")
c_bar = plt.colorbar()
c_bar.set_label("Temp V/S icecreame sales")
plt.show()
```

```python
import numpy as np
size = np.random.randint(low=500,high=2000,size=7)
size
```

```python
plt.scatter(temp,icecream_sales,c=color,s=size,marker='<',cmap="Oranges",vmin=10,vmax=20)
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")
c_bar = plt.colorbar()
c_bar.set_label("Temp V/S icecreame sales")
plt.show()
```

# Draw Two Scatter Plots In One Figure

```python
plt.scatter(temp,icecream_sales,c='y',marker='<',s=100,edgecolor='k',linewidth=1)

plt.scatter(temp,tea_sales,c='g',marker='<',s=100,edgecolor='k',linewidth=1)
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")

plt.show()
```

```python
plt.scatter(temp,icecream_sales,c='y',marker='<',s=100,edgecolor='k',linewidth=1,label="icecream sales")

plt.scatter(temp,tea_sales,c='g',marker='<',s=100,edgecolor='k',linewidth=1,label="tea sales")
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,3),facecolor="lightblue",linewidth=2,edgecolor="k")

plt.scatter(temp,icecream_sales,c='y',marker='<',s=100,edgecolor='k',linewidth=1,label="icecream sales")

plt.scatter(temp,tea_sales,c='g',marker='<',s=100,edgecolor='k',linewidth=1,label="tea sales")
plt.title("Temprature V/S icecreame sales")
plt.xlabel("temprature")
plt.ylabel("icecream sales")
plt.legend()
plt.show()
```
